### Importação dos módulos e da tabela

```python
import pandas as pd

tabela = pd.read_csv("exportacao_full.csv")
display(tabela)
```

### Filtragem dos dados e informações gerais

```python
#tabela = tabela.loc[linha, coluna]
tabela = tabela.loc[tabela["Country"] == "France", :] # o : indica que você quer pegar todos os valores
tabela = tabela.loc[tabela["Year"] >= 2016, :]
tabela.to_csv('exportacao_france.csv', index=False)
```

```python
tabela = pd.read_csv("exportacao_france.csv")
display(tabela)
display(tabela.info())
```

### Como foi a evolução das exportações para a França ao longo dos anos

```python
tabela_anos_exportacao = tabela[["Year", "US$ FOB"]].groupby("Year").sum()

def formatar_valor(valor):
    valor_formatado = f"US${valor:,}"
    return valor_formatado

tabela_anos_exportacao["US$ FOB"] = tabela_anos_exportacao["US$ FOB"].apply(formatar_valor)

display(tabela_anos_exportacao)
```

### Quais os produtos mais exportados ao longo de todo período?

```python
tabela_produtos = tabela[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum()
tabela_produtos = tabela_produtos.sort_values(by="US$ FOB", ascending=False)
tabela_produtos["US$ FOB"] = tabela_produtos["US$ FOB"].apply(formatar_valor)

display(tabela_produtos)
```

### Em 2020, qual cidade mais exportou para a França?

```python
tabela_2020 = tabela.loc[tabela["Year"] == 2020, :]
tabela_2020_cidades = tabela_2020[["City", "US$ FOB"]].groupby("City").sum()
tabela_2020_cidades = tabela_2020_cidades.sort_values(by="US$ FOB", ascending=False)
tabela_2020_cidades["US$ FOB"] = tabela_2020_cidades["US$ FOB"].apply(formatar_valor)

display(tabela_2020_cidades)
```

### O que as duas maiores cidades exportaram tanto?

```python
cidade = tabela_2020_cidades.index[0] # Para trocar de cidade basta trocar o índice para 0 (1° maior) ou 1 (2° maior)

tabela_cidade = tabela_2020.loc[tabela_2020["City"] == cidade]
tabela_produtos_cidade = tabela_cidade[["SH2 Description", "US$ FOB"]].groupby("SH2 Description").sum()
tabela_produtos_cidade = tabela_produtos_cidade.sort_values(by="US$ FOB", ascending=False)
tabela_produtos_cidade["US$ FOB"] = tabela_produtos_cidade["US$ FOB"].apply(formatar_valor)

print(f"Produtos mais exportados da cidade: {cidade}")
display(tabela_produtos_cidade)
```
